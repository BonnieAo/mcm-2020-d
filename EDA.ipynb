# Data Descriptions

1) `matches.csv`

- `MatchID`: A unqiue identifier for each match played during the season, and reflects the order of the match in the season.
- `OpponentID`: A unqiue identifier for the opposing team played in the match. **Note that the Huskies play each opposing team twice during the season**
- `Outcome`: Result of the match, eiter `win`, `loss`, or `tie`.
- `OwnScore`: Number of goals scored by the Huskies.
- `OpponentScore`: Number of goals scored by the Opposing Team.
- `Side`: Whether the Huskies were the `home` team or `away` team.
- `CoachID`: A unqiue identifier for the Huskies coach for this match.

2) `passingevents.csv`

- `MatchID`: A unqiue identifier for each match played during the season (see `matches.csv`).
- `TeamID`: A unqiue identifier for the team involved in the pass (either `Huskies` or `OpponentID` from `matches.csv`).
- `OriginPlayerID`: A unqiue identifier for the Player at the origin of the pass. The `PlayerID` has the form `TeamID_PlayerPosition##` where `TeamID` denotes the team on which the player plays and `PlayerPosition` reflects the player's position. Possible positions are: `F`:forward, `D`:defense, `M`:midfield, or `G`:goalkeeper
- `DestinationPlayerID`: A unqiue identifier for the Player at the destination of the pass. (see `OriginPlayerID`)
- `MatchPeriod`: The half in which the event took place. `1H`: first half, `2H`: second half
- `EventTime`: The time in seconds during the `MatchPeriod` (1st or 2nd half) at which the event took place.
- `EventSubType`: The type of pass made. Can be one of: `Head pass`, `Simple pass`, `Launch`, `High pass`, `Hand pass`, `Smart pass`, `Cross`.
- `EventOrigin_x`: The x-coordinate on the field at which the pass originated. The x-coordinate is in the range $[0, 100]$ and is **oriented from the perspective of the attacking team**, where $0$ indicates the team's own goal, and $100$ indicates the oppositing team's goal.
- `EventOrigin_y`: The y-coordinate on the field at which the pass originated. The y-coordinate is in the range $[0, 100]$ and is **oriented from the perspective of the attacking team**, where $0$ indicates the team's left-hand side, and $100$ indicates the team's right-hand side.
- `EventDestination_x`: The x-coordinate on the field at the pass destination. (see `EventOrigin_x`)
- `EventDestination_y`: The y-coordinate on the field at the pass destination.  (see `EventOrigin_y`)

3) `fullevents.csv`

- `MatchID`: A unqiue identifier for each match played during the season (see `matches.csv`).

# TODO: Create a network for the ball passing between players

```python
huskies_passes = passings_df[passings_df.TeamID == 'Huskies']
```

```python
class Player:
    def __init__(self, player_id: str):
        self.id = player_id
        pos_n = player_id.split('_')[1] # [F|DM|G] + one or two digit(s)
        self.pos = pos_n[0] # F:forward, D:defense, M:midfield, or G:goalkeeper
        self.n = int(pos_n[1:]) # n-th player in the same position
        self.coord = [0, 0] # football field, left is our goal, right is opponent's goal,
                            # x in [0, 100], y in [0, 100]
        if self.pos == 'F':
            self.coord[0] = 90
        elif self.pos == 'M':
            self.coord[0] = 60
        elif self.pos == 'D':
            self.coord[0] = 30
        elif self.pos == 'G':
            self.coord[0] = 0
        else:
            raise NameError("Invalid player position")
        
        self.coord[1] = (self.n / 10) * 100
    
    def __hash__(self):
        return self.player_id.__hash__()
```

```python
from typing import List, Tuple
import matplotlib.lines as mlines

class Network:
    def __init__(self, players: List[Player]):
        self.players = { n.id: n for n in players }
        self.links: dict  = {} # '<player_id1>,<player_id2>':<count>
    
    def add_passing(self, player1: str, player2: str):
        pair = player1 + ',' + player2 # str __hash__
        if pair in self.links:
            self.links[pair] += 1
        else:
            self.links[pair] = 1
    
    def to_graph(self):
        # draw all players as points
        for i, p in self.players.items():
            c = p.coord
            plt.plot([c[0]], [c[1]], marker='o', color='r')
            
        # draw links
        for i1, p1 in self.players.items():
            for i2, p2 in self.players.items():
                pair = i1 + ',' + i2
                count = self.links.get(pair, 0)
                
                coord1 = p1.coord
                coord2 = p2.coord
                
                max_pass_count = max(self.links.values())
                min_pass_count = min(self.links.values())
                
                ax = plt.gca()
                # use color to represent count
                l = mlines.Line2D([coord1[0], coord2[0]], [coord1[1], coord2[1]],
                                  color=plt.get_cmap('rainbow')(
                                      (count - min_pass_count) / (max_pass_count - min_pass_count)
                                  )
                                 )
                ax.add_line(l)
        plt.colorbar()
        plt.show()

    def pass_hotspot(self):
        """Visualize the number of passes a player initiated"""
        player_passes = {}
        for p in self.links:
            i1, _ = p.split(',')
            if i1 in player_passes:
                player_passes[i1] += 1
            else:
                player_passes[i1] = 1
        max_pass_count = max(player_passes.values())
        min_pass_count = min(player_passes.values())
        cmap = plt.get_cmap('rainbow')
        fig, ax = plt.subplots(1, 1)
        for i, p in self.players.items():
            c = p.coord
            count = player_passes.get(i, 0)
            ax.plot([c[0]], [c[1]], marker='o', color=cmap(
                                      (count - min_pass_count) / (max_pass_count - min_pass_count)
                                  ))
        bar = mpl.cm.ScalarMappable(cmap=cmap)
        fig.colorbar(bar)
        fig.tight_layout()
        plt.show()
```

```python
player_ids = list(np.unique(pd.concat([huskies_passes['OriginPlayerID'], huskies_passes['DestinationPlayerID']])))
players = [Player(pi) for pi in player_ids]
network = Network(players)
for i, row in huskies_passes.iterrows():
    network.add_passing(row['OriginPlayerID'], row['DestinationPlayerID'])
# network.to_graph()
```

# How passes affect scores

```python
def get_ownscore(data, match_id: int):
    return data[data['MatchID'] == match_id]['OwnScore'].tolist()[0]

def get_score_diff(data, match_id: int):
    own = data[data['MatchID'] == match_id]['OwnScore'].tolist()[0]
    opponent = data[data['MatchID'] == match_id]['OpponentScore'].tolist()[0]
    return own - opponent

def get_outcome(data, match_id: int):
    outcome = data[data['MatchID'] == match_id]['Outcome'].tolist()[0]
    if outcome == 'win':
        return 1
    elif outcome == 'tie':
        return 0
    elif outcome == 'lose':
        return -1
```

```python
def get_total_n_passes(data, match_id: int):
    return data[data['MatchID'] == match_id].shape[0]

def plot_total_n_passes_against(x_func, y_func, matches_data=matches_df, passes_data=huskies_passes, ax=plt):
    match_ids = list(np.unique(passes_data['MatchID']))
    passes = [x_func(passes_data, mi) for mi in match_ids]
    outcomes = [y_func(matches_data, mi) for mi in match_ids]
    ax.scatter(passes, outcomes)
```

```python
plot_total_n_passes_against(get_total_n_passes, get_outcome, ax=plt)
plt.title("Total number of passes in a match vs. outcomes")
plt.xlabel("Number of passes in a match")
plt.ylabel("Outcomes")
```

```python
fig, axs = plt.subplots(2)

plot_total_n_passes_against(get_total_n_passes, get_ownscore, ax=axs[0])
axs[0].set_title("Total number of passes in a match vs. scores")
axs[0].set_xlabel("Number of passes in a match")
axs[0].set_ylabel("Scores")

plot_total_n_passes_against(get_total_n_passes, get_score_diff, ax=axs[1])
axs[1].set_title("Total number of passes in a match vs. score differences")
axs[1].set_xlabel("Number of passes in a match")
axs[1].set_ylabel("Score diff")

fig.tight_layout()

plt.show()
```
